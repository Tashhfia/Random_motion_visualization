## Imports

```python
import matplotlib.pyplot as plt

from random import choice
```

```python
class random_motion:
    
    def __init__(self, num_points = 5000):
        
        """Initialize attributes of a walk."""
        self.num_points = num_points
        self.x_cord = [0]
        self.y_cord = [0]
    
    def motion(self):
        """ a method to determine the all points along the walk"""
        
        while (len(self.x_cord) < self.num_points):
            
            # steps taken along x 
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_movement = x_direction * x_distance
            
            # steps taken along y
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_movement = y_direction * y_distance
            
            # reject steps that go nowhere
            if x_movement == 0 and y_movement == 0:
                continue
                
            # get the new cordinates
            self.x_cord.append(self.x_cord[-1] + x_movement)
            self.y_cord.append(self.y_cord[-1] + y_movement)    
```

## Visualizing the walk

```python
pollen_1 = random_motion(50_000)
pollen_1.motion()

```

```python
plt.style.use('seaborn-poster')

fig, ax = plt.subplots()
point_numbers = range(pollen_1.num_points)
ax.set_title('Random motion of a pollen visualized')
ax.scatter(pollen_1.x_cord, pollen_1.y_cord, c = point_numbers, cmap = plt.cm.autumn, s = 10, edgecolors= 'none' )
```

## Marking the starting and ending points

```python
fig, ax = plt.subplots()
point_numbers = range(pollen_1.num_points)
ax.set_title('Random motion of a pollen visualized')
ax.scatter(pollen_1.x_cord, pollen_1.y_cord, c = point_numbers, cmap = plt.cm.autumn, s = 10, edgecolors= 'none' )
# mark start and end
ax.scatter(0,0, c = 'blue', s = 100)
ax.scatter(pollen_1.x_cord[-1], pollen_1.y_cord[-1], c = 'green', s = 100)
# removing axes
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
```
